```python
from matplotlib import pyplot as plt
%matplotlib inline
```

# Pie Chart

```python
areas = ['Marketing', 'Sales', 'Development', 'HR', 'Customer Support']
budget = [2.5, 4, 10, 1, 6]
```

```python
plt.pie(budget, labels=areas, autopct='%0.1f%%')
plt.show()
```

# Scatter Plot

```python
x = [4, 6, 9, 10, 50, 11, 22, 99]
y = [7, 4, 5, 8, 1, 2, 6, 3]
```

```python
plt.scatter(x, y)
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Age V/s Salary Graph')
plt.show()
```

# 2-D Graph

```python
product = [1, 2, 3, 4, 5, 6, 7, 8]
online_price = [233, 456, 770, 120, 222, 444, 675, 399]
offline_price = [400, 300, 299, 600, 1000, 199, 99, 499]
sikkim_price = [200, 300, 599, 100, 1000, 699, 999, 499]
```

```python
plt.plot(product, online_price, color='g', marker='d', markersize = 20)
plt.xlabel('Product')
plt.ylabel('Prices')
plt.title('Product Price Comparision')
plt.show()
```

```python
plt.plot(product, online_price, color='g', marker='d', markersize = 10, label='Online Prices')
plt.plot(product, offline_price, color='b', marker='^', markersize = 10, label='Offline Prices')
plt.plot(product, sikkim_price, color='r', marker='*', markersize = 10, label='Sikkim Pirice')
plt.xlabel('Product')
plt.ylabel('Prices')
plt.title('Product Price Comparision')
plt.legend()
plt.show()
```

# Bar Graph

### Displays the quantity

```python
year = [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
price = [65000, 69000, 75000, 55000, 52000, 64000, 81000, 85000, 89000, 65000, 71000]
```

```python
plt.bar(year, price, color='purple')
# plt.barh(year, price, color='purple')
plt.ylabel('Price in $')
plt.xlabel('Year')
plt.title('Year by Year Price Comp.')
plt.show()
```
